```python
import matplotlib.pyplot as plt
import matplotlib.animation as anim
import numpy as np

def uzupelnij_plik():
    f = open("dane.txt","w")
    w = str(list(range(100))).replace(",","").replace("[","").replace("]","")
    print(w)
    f.write(w)
    f.close()

def odczytaj():
    f = open("dane.txt","r")
    w = f.read().split(" ")
    y = []
    for el in w:            # y = [float(el) for el in w]
        y.append(float(el)) #
    f.close()
    return y

def generuj():
    f = open("dane.txt","a")
    f.write(" "+str(np.random.uniform(0,50)))
    f.close()

uzupelnij_plik()
yvalues = odczytaj()
fig = plt.figure(1)
lines = plt.plot(yvalues)
# plt.xlabel("dane")

def update(i):
    generuj()
    y = odczytaj()
    plt.clf()
    plt.plot(y)
    # lines[0].set_xdata(range(len(y)))
    # lines[0].set_ydata(y)
    # return lines

ani = anim.FuncAnimation(fig, update, interval=100)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as anim
import mpl_toolkits.mplot3d

x = np.linspace(0,10,100)
y = np.linspace(0,10,100)

X, Y = np.meshgrid(x, y)
Z = np.cos(X)+np.sin(Y)

global xr, yr, zr

fig = plt.figure()
ax = plt.axes(projection='3d')
xr = np.array(np.random.uniform(0,10))
yr = np.array(np.random.uniform(0,10))
zr = np.array(np.cos(xr) + np.sin(yr))
ax.scatter(xr,yr,zr,color='red')
ax.plot_wireframe(X,Y,Z)

def update(i):
    global xr, yr, zr
    nx, ny = np.random.uniform(0,10), np.random.uniform(0,10)
    xr=np.append(xr,nx)
    yr=np.append(yr,ny)
    zr=np.append(zr,np.cos(nx)+np.sin(ny))
    ax.cla()
    ax.scatter(xr,yr,zr,color='red')
    ax.plot(xr,yr,zr,color='green')
    ax.plot_wireframe(X,Y,Z)

ani = anim.FuncAnimation(fig, update)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as anim
import mpl_toolkits.mplot3d

x = np.linspace(0,10,100)
y = np.linspace(0,10,100)

X, Y = np.meshgrid(x, y)
Z = np.cos(X)+np.sin(Y)

global xr, yr, zr

fig = plt.figure()
ax = plt.axes(projection='3d')
xr = np.array(np.random.uniform(0,10))
yr = np.array(np.random.uniform(0,10))
zr = np.array(np.cos(xr) + np.sin(yr))
ax.scatter(xr,yr,zr,color='red')
ax.plot_wireframe(X,Y,Z)

def update(i):
    global xr, yr, zr
    nx, ny = np.random.uniform(0,10), np.random.uniform(0,10)
    nz = np.cos(nx)+np.sin(ny)
    if nz<zr.min():
        xr=np.append(xr,nx)
        yr=np.append(yr,ny)
        zr=np.append(zr,nz)
    ax.cla()
    ax.scatter(xr,yr,zr,color='red')
    ax.plot(xr,yr,zr,color='green')
    ax.plot_wireframe(X,Y,Z)

ani = anim.FuncAnimation(fig, update)
plt.show()
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

jablka = cv2.imread("jablka.jpeg")
pomarancze = cv2.imread("orange.jpg")

jablka2 = jablka[:500,:800,:]
pomarancze2 = pomarancze[:500,:800,:]

pomarancze2[1:150,250:300,:] = pomarancze2[101:250,650:700,:]

blend = cv2.add(jablka2,pomarancze2)
blend2 = cv2.addWeighted(jablka2,1,pomarancze2,0.1,0)
cv2.putText(blend, "Jablka", (50,50), cv2.FONT_HERSHEY_SIMPLEX, 2, 255)
cv2.imshow("Jablka", jablka2)
cv2.imshow("Pomarancze", pomarancze2)
cv2.imshow("Blend", blend)
cv2.imshow("Blend2", blend2)
cv2.waitKey(0)
cv2.destroyAllWindows()

print("Rozmiar jablek: ",jablka.shape, "rozmiar pomaranczy:",pomarancze.shape)
```
